```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
df=pd.read_csv(r'/Users/praveenmadhogaria/Desktop/Desktop - Praveen’s MacBook Pro/CSV:Data files/movies.csv')
df
```

```python
df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print(col,pct_missing,'%')
```

```python
df.dtypes
```

```python
# Replacing Nan values with 0 so datatypes can be changed

df=df.fillna(0)
```

```python
df['gross']=df['gross'].astype('int64')
```

```python
df['budget']=df['budget'].astype('int64')
```

```python
df.dtypes
```

```python
df.drop_duplicates()
```

```python
#creating scatter plot to find correlation between budget and gross column

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earning')
plt.show()
```

```python
df.sort_values(by=['gross'],ascending=False,inplace=True)
df
```

```python
#budget vs Gross plotting using seaborn

sb.regplot(data=df,x='budget',y='gross',scatter_kws={"color":"black"},line_kws={"color":"red"})
```

```python
# Generating correlation matrix

df.corr(numeric_only=True)     #'pearson', 'kendall', 'spearman' - correlation methods
```

```python
# Generating heatmap

corr_matrix=df.corr(numeric_only=True)

sb.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
#Converting non-numeric values to distinct numeric values in order to find correlation.

df_factorized=df.apply(lambda x: x.factorize(sort = True)[0])
df_factorized
```

```python
df
```

```python
df_factorized.corr()
```

```python
# correlation matrix of all columns including non-numeric

corr_matrix=df_factorized.corr(numeric_only=True)

#specify size of heatmap
fig, ax = plt.subplots(figsize=(10, 5))

sb.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```
