```python
import pandas as pd

df = pd.read_excel('/Users/praveenmadhogaria/Desktop/Pandas_Tutorial/Customer Call List.xlsx')
df
```

```python
df.drop_duplicates(inplace=True)    #dropping duplicates
df
```

```python
df.drop(columns='Not_Useful_Column', inplace=True)         #dropping specific column
df
```

```python
df['Last_Name']=df['Last_Name'].str.strip("/..._ ")
df
```

```python
df['Phone_Number']=df['Phone_Number'].replace(to_replace='[^0-9]',value='',regex=True)
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))  #converting column to string
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:])   #formatting phone number
df
```

```python
#df['Phone_Number'].replace(to_replace='nan--',value='',regex=True,inplace=True)   #deleting nan values
df['Phone_Number'].replace(to_replace='--',value='',regex=True,inplace=True)       #deleting -- values
df
```

```python
df[['Street Name','State','Zipcode']]=df['Address'].str.split(',',expand=True)
df
```

```python
df['Paying Customer'].replace(to_replace='Yes',value='Y',inplace=True)   #replacing yes to Y
#df['Paying Customer'].replace(to_replace='No',value='N',inplace=True)   #replacing no to N
df
```

```python
#df['Do_Not_Contact'].replace(to_replace='Yes',value='Y',inplace=True)   #replacing yes to Y
df['Do_Not_Contact'].replace(to_replace='No',value='N',inplace=True)   #replacing no to N
df
```

```python
df.fillna(value='',inplace=True)   #filling Nan with approriate values
df
```

```python
df.replace(to_replace='N/a',value='',inplace=True)      #replacing Na with blanks
df
```

```python
for x in df.index:                                      #if Do_Not_Contact is Y delete row
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x,inplace=True)
        
        
df
```
